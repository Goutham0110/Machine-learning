```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model as L
from sklearn.model_selection import train_test_split
```

```python
#importing dataset
dataset='https://raw.githubusercontent.com/Goutham0110/Students-Performance/main/student-mat.csv'
df=pd.read_csv(dataset,sep=';')
df.corr() #to find which variable has most impact on the final marks(i.e. close to 1.00)
```

```python
#pre-processing variables
def percent(x):
  y=(x-x.min())/(x.max()-x.min())
  return y*100

X=percent(df[['G1']])
Y=percent(df[['G3']])
```

```python
#Execution of Linear regression model
x_train,x_test,y_train,y_test= train_test_split(X,Y,test_size=0.15,random_state=3)
lm=L.LinearRegression()
lm.fit(x_train,y_train)
yhat=lm.predict(x_test)
```

```python
#visulizing the data and the model
plt.scatter(x_train,y_train, color='cyan')
plt.plot(x_test,yhat, linewidth=3)
plt.show()
```
